```python
import os
import xml.etree.ElementTree as ET
from PIL import Image

# Thư mục chứa ảnh gốc và annotation
image_dir = '/home/jupyter-iec_thicao/Dataset/SVRDD_VOC/Xicheng/JPEGImages'
annotation_dir = '/home/jupyter-iec_thicao/Dataset/SVRDD_VOC/Xicheng/Annotations'
output_root = '/home/jupyter-iec_thicao/Dataset/SVRDD_VOC/CroppedObjects'

os.makedirs(output_root, exist_ok=True)

# Duyệt qua từng file annotation
for xml_file in os.listdir(annotation_dir):
    if not xml_file.endswith('.xml'):
        continue

    tree = ET.parse(os.path.join(annotation_dir, xml_file))
    root = tree.getroot()

    filename = root.find('filename').text
    image_path = os.path.join(image_dir, filename)

    if not os.path.exists(image_path):
        print(f"Không tìm thấy ảnh: {filename}")
        continue

    image = Image.open(image_path)
    base_name = os.path.splitext(filename)[0]

    for i, obj in enumerate(root.findall('object')):
        label = obj.find('name').text

        bndbox = obj.find('bndbox')
        xmin = int(float(bndbox.find('xmin').text))
        ymin = int(float(bndbox.find('ymin').text))
        xmax = int(float(bndbox.find('xmax').text))
        ymax = int(float(bndbox.find('ymax').text))

        # Cắt object
        cropped = image.crop((xmin, ymin, xmax, ymax))

        # Tạo thư mục nhãn nếu chưa có
        label_dir = os.path.join(output_root, label)
        os.makedirs(label_dir, exist_ok=True)

        # Lưu ảnh đã cắt
        out_filename = f"{base_name}_{i}.jpg"
        cropped.save(os.path.join(label_dir, out_filename))
        print(f"Đã lưu: {os.path.join(label_dir, out_filename)}")
```

```python
import os
import random
import shutil
#
# Đường dẫn thư mục
train_dir = '/home/jupyter-iec_thicao/Dataset/SVRDD_VOC/Train'
test_dir = '/home/jupyter-iec_thicao/Dataset/SVRDD_VOC/Test'


for class_name in os.listdir(train_dir):
    os.makedirs(f"{test_dir}/{class_name}", exist_ok=True)
    image_files = [f for f in os.listdir(f"{train_dir}/{class_name}") if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    num_to_move = int(0.2 * len(image_files))
    images_to_move = random.sample(image_files, num_to_move)
    for img in images_to_move:
        src = os.path.join(f"{train_dir}/{class_name}", img)
        dst = os.path.join(f"{test_dir}/{class_name}", img)
        shutil.move(src, dst)
    print(f"Đã chuyển {len(images_to_move)} ảnh từ {train_dir}/{class_name} sang {test_dir}/{class_name}.")
```
