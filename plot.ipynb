```python
import matplotlib.pyplot as plt

# Danh sách các mô hình
models = [
    'ResNet18', 'ResNet34', 'ResNet50', 'EfficientNet B3',
    'EfficientViT b1', 'EfficientViT D2', 'EfficientViT D3', 'ResEViT_road'
]

# FPS tương ứng với từng model
raspberry_pi_4 = [1.2, 1.3, 1.4, 2.5, 1.6, 1.2, 1.1, 1.3]  # Giả định
jetson_nano = [32, 18, 11, 12, 16, 10, 6, 17]               # Giả định

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.plot(models, raspberry_pi_4, label='Raspberry PI 4', color='blue', linewidth=2, marker='o')
plt.plot(models, jetson_nano, label='Jetson Nano', color='red', linewidth=2, marker='o')

# Tùy chỉnh kiểu hiển thị
plt.xlabel('Model', fontsize=16, fontweight='bold')
plt.ylabel('FPS', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)  # tăng kích cỡ chữ chú thích
plt.legend()
plt.grid(True)

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```
